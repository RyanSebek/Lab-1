```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv("house_prices.csv")
df.head
```

```python
df['Balcony']=df['Balcony'].fillna(0)
df['beds'] = df['Title'].str.extract(r'(\d+)\s*BHK').astype(float)
df_clean = df.drop(columns=['Index','Dimensions','Plot Area','facing','overlooking','Society','Super Area','Status','Floor','Description','Title'])
```

```python
df_clean.head()
```

```python
df_clean['Carpet Area'] = df_clean['Carpet Area'].str.replace(' sqft','',regex=False)
df_clean['Carpet Area'] = pd.to_numeric(df_clean['Carpet Area'], errors='coerce')
```

```python
df_clean.head()
```

```python
def convert_amount(val):
    if isinstance(val, str):
        val = val.strip()
        if 'Cr' in val:
            return float(val.replace('Cr', '').strip()) * 1e7
        elif 'Lac' in val:
            return float(val.replace('Lac', '').strip()) * 1e5
    return np.nan

df_clean['Amount(in rupees)'] = df_clean['Amount(in rupees)'].apply(convert_amount)
```

```python
df_clean['Bathroom'] = pd.to_numeric(df_clean['Bathroom'], errors='coerce')
df_clean['Balcony'] = pd.to_numeric(df_clean['Balcony'], errors='coerce')
```

```python
df_clean = df_clean.dropna(subset=['Price (in rupees)'])
```

```python
df_clean.head()
```

```python
df_clean['Furnishing'] = df_clean['Furnishing'].replace({
    'Furnished': 3,
    'Semi-Furnished': 2,
    'Unfurnished': 1
}).astype(float)

df_clean['Ownership'] = df_clean['Ownership'].replace({
    'Freehold': 4,
    'Co-operative Society': 3,
    'Power Of Attorney': 2,
    'Leasehold': 1,
}).fillna(0).astype(float)

df_clean['Car Parking'] = df_clean['Car Parking'].fillna('None').astype(str)
```

```python
df_clean.head()
```

```python
df_model = df_clean.dropna()
X = df_model.drop(columns=['Price (in rupees)'])
y = df_model['Price (in rupees)']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.2, random_state=42)
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
y_pred_lr = lr_model.predict(X_test)
```

```python
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train, y_train)
y_pred_ridge = ridge_model.predict(X_test)
```

```python
mse_lr = mean_squared_error(y_test, y_pred_lr)
rmse_lr = mse_lr**0.5

mse_ridge = mean_squared_error(y_test, y_pred_ridge)
rmse_ridge = mse_ridge**0.5

mse_lr, rmse_lr, mse_ridge, rmse_ridge
```

```python
print("Linear Regression Root Mean Squared Error:", rmse_lr)
print("Ridge Regression Root Mean Squared Error:", rmse_ridge)
```
